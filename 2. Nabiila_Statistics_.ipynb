# Step 1: Data Preprocessing

```
#import the library, read file from github, dataoverview
import pandas as pd                                                                                                  
df_raw1 = pd.read_csv("https://raw.githubusercontent.com/Syukrondzeko/Fellowship-5/main/Seasons_Stats.csv")          
df_raw1                                                                                                             
```

There are 24.691 rows and 53 column in this data type.

```
#filter data with year 2017 only
df_raw2 = df_raw1[df_raw1['Year'] == 2017]                         
df_raw2
```

The rows decreasing from 24.691 to 595 rows

```
#check data type
df_raw2.dtypes
```

The date type has been match. We donâ€™t need to do data transformation

```
#Check duplicate values
df_raw2.duplicated(['Player'])
```

'True' indicates there are duplicate values

```
#remove duplicate values
df_raw3 = df_raw2.drop_duplicates(subset=['Player'])  
df_raw3  
```

The rows decrease from 595 to 486 rows.

```
#check the null column
df_raw3.info()   
```

0 indicates there is no value on the column. We will drop them.

```
#drop null columns
df = df_raw3.drop(columns=['blanl', 'blank2'])                
df
```

The columns decrease from 53 to 51

We have done our data reprocessing.
After this reprocessing, the data rows are 486 and columns is 51.

# Step 2: Answering Question

## 1. Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?


```
#Chi-Square Normality test
from scipy import stats
import scipy as scipy
stats.chisquare(df['Age'])
```

If the p-value> 0.05, the distribution of the data is normal. Above we have 0.9 **The data distribution is normal**

```
df[["Tm", "Player", "Age"]].head() 
```

```
#fing the youngest player on every team
youngest = df.groupby('Tm')['Age', 'Player'].agg(min)
youngest
```

```
#find the oldest player on every team.
oldest = df.groupby('Tm')['Age', 'Player'].agg(max)
oldest
```

```
#optinal. Join the data
no1 = pd.merge(youngest, oldest, on='Tm', how = 'inner')
no1.head()
```

```
#rename the column name
no1.rename(columns ={'Age_x' : 'YoungestAge', 'Age_y' : 'OldestAge', 'Player_x': 'YoungestPlayer', 'Player_y' : 'OldestPlayer'}, inplace=True )
no1.head()
```

## 2. Which player has the most minutes played (MP) in each position (Pos)?

```
#shapiro vilk test (normality test)
stats.shapiro(df['MP'])
```

We see the p-value. The p-value is the second index of the result above. The normal data has p-value 0> 0.05

the p-value of the data is 3.89. **The data has normal distirbution**

```
#player with most minutes played on every position
no2 = df.groupby('Pos')['MP', 'Player'].agg(max)
no2
```

## 3. Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?


### TRB%

```
#team with highest average of TRB%
trb = df.groupby(['Tm']).mean()['TRB%'].sort_values(ascending=False)
trb.head(1)
```

Team with highest TRB% is WAS

### AST%

```
#team with highest average of AST%
ast = df.groupby(['Tm']).mean()['AST%'].sort_values(ascending=False)
ast.head(1)
```

Team with highest AST% is DEN

### STL%

```
#team with highest average of STL%
stl = df.groupby(['Tm']).mean()['STL%'].sort_values(ascending=False)
stl.head(1)
```

### BLK%

```
#Team with highest average of BLK%
blk = df.groupby(['Tm']).mean()['BLK%'].sort_values(ascending=False)
blk.head(1)
```

Team with highest BLK% is GSW
